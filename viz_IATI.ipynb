```python
import pandas as pd
from datetime import datetime
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('datasets/IATI_data.csv')
```

```python
df = df[df['day_length']!=0]
```

```python
sectors = ['Water Supply & Sanitation']
bysect = df[df['sector_group'].isin(sectors)]
cholera = df[df['description'].astype(str).str.lower().str.contains('cholera')]
cholera_vax = df[(df['description'].astype(str).str.lower().str.contains('cholera')) & 
             (df['description'].astype(str).str.lower().str.contains('vaccin'))]
```

```python
latest_start = pd.to_datetime(df['day_start']).apply(lambda x: max(x, datetime(2017, 1, 1)))
earliest_end = pd.to_datetime(df['day_end']).apply(lambda x: min(x, datetime(2017, 12, 31)))
df['2017_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
latest_start = pd.to_datetime(df['day_start']).apply(lambda x: max(x, datetime(2018, 1, 1)))
earliest_end = pd.to_datetime(df['day_end']).apply(lambda x: min(x, datetime(2018, 12, 31)))
df['2018_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
df['daily_budget'] = df['commitment']/df['day_length']
df['2017_budget'] = df['daily_budget'] * df['2017_days']
df['2018_budget'] = df['daily_budget'] * df['2018_days']
df['daily_spend'] = df['spend']/df['day_length']
df['2017_spend'] = df['daily_spend'] * df['2017_days']
df['2018_spend'] = df['daily_spend'] * df['2018_days']
#np.min(pd.to_datetime(bysect['day_end']), datetime(year=2017, month=12, day=31))
```

```python
latest_start = pd.to_datetime(bysect['day_start']).apply(lambda x: max(x, datetime(2017, 1, 1)))
earliest_end = pd.to_datetime(bysect['day_end']).apply(lambda x: min(x, datetime(2017, 12, 31)))
bysect['2017_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
latest_start = pd.to_datetime(bysect['day_start']).apply(lambda x: max(x, datetime(2018, 1, 1)))
earliest_end = pd.to_datetime(bysect['day_end']).apply(lambda x: min(x, datetime(2018, 12, 31)))
bysect['2018_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
bysect['daily_budget'] = bysect['commitment']/bysect['day_length']
bysect['2017_budget'] = bysect['daily_budget'] * bysect['2017_days']
bysect['2018_budget'] = bysect['daily_budget'] * bysect['2018_days']
bysect['daily_spend'] = bysect['spend']/bysect['day_length']
bysect['2017_spend'] = bysect['daily_spend'] * bysect['2017_days']
bysect['2018_spend'] = bysect['daily_spend'] * bysect['2018_days']
#np.min(pd.to_datetime(bysect['day_end']), datetime(year=2017, month=12, day=31))
```

```python
latest_start = pd.to_datetime(cholera['day_start']).apply(lambda x: max(x, datetime(2017, 1, 1)))
earliest_end = pd.to_datetime(cholera['day_end']).apply(lambda x: min(x, datetime(2017, 12, 31)))
cholera['2017_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
latest_start = pd.to_datetime(cholera['day_start']).apply(lambda x: max(x, datetime(2018, 1, 1)))
earliest_end = pd.to_datetime(cholera['day_end']).apply(lambda x: min(x, datetime(2018, 12, 31)))
cholera['2018_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
cholera['daily_budget'] = cholera['commitment']/cholera['day_length']
cholera['2017_budget'] = cholera['daily_budget'] * cholera['2017_days']
cholera['2018_budget'] = cholera['daily_budget'] * cholera['2018_days']
cholera['daily_spend'] = cholera['spend']/cholera['day_length']
cholera['2017_spend'] = cholera['daily_spend'] * cholera['2017_days']
cholera['2018_spend'] = cholera['daily_spend'] * cholera['2018_days']
```

```python
latest_start = pd.to_datetime(cholera_vax['day_start']).apply(lambda x: max(x, datetime(2017, 1, 1)))
earliest_end = pd.to_datetime(cholera_vax['day_end']).apply(lambda x: min(x, datetime(2017, 12, 31)))
cholera_vax['2017_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
latest_start = pd.to_datetime(cholera_vax['day_start']).apply(lambda x: max(x, datetime(2018, 1, 1)))
earliest_end = pd.to_datetime(cholera_vax['day_end']).apply(lambda x: min(x, datetime(2018, 12, 31)))
cholera_vax['2018_days'] = (earliest_end - latest_start).apply(lambda x: max(0, x.days + 1))
cholera_vax['daily_budget'] = cholera_vax['commitment']/cholera_vax['day_length']
cholera_vax['2017_budget'] = cholera_vax['daily_budget'] * cholera_vax['2017_days']
cholera_vax['2018_budget'] = cholera_vax['daily_budget'] * cholera_vax['2018_days']
cholera_vax['daily_spend'] = cholera_vax['spend']/cholera_vax['day_length']
cholera_vax['2017_spend'] = cholera_vax['daily_spend'] * cholera_vax['2017_days']
cholera_vax['2018_spend'] = cholera_vax['daily_spend'] * cholera_vax['2018_days']
```

```python
print(f'By sector tot daily budget \n\t2017: {round(bysect["2017_budget"].sum() / 365)} USD \n\t2018: {round(bysect["2018_budget"].sum() / 365)} USD')
print(f'Cholera tot daily budget \n\t2017: {round(cholera["2017_budget"].sum() / 365)} USD \n\t2018: {round(cholera["2018_budget"].sum() / 365)} USD')
print(f'Cholera + vaccine tot daily budget \n\t2017: {round(cholera_vax["2017_budget"].sum() / 365)} USD \n\t2018: {round(cholera_vax["2018_budget"].sum() / 365)} USD')
```

```python
days_2017 = pd.date_range(datetime(2017,1,1), datetime(2017,12,31), periods=365).tolist()
days_2018 = pd.date_range(datetime(2018,1,1), datetime(2018,12,31), periods=365).tolist()
```

```python
newdf = cholera[['aid', 'funder_ref', 'daily_spend']]
for d in days_2017 + days_2018:
    is_after_start = pd.to_datetime(cholera['day_start']).apply(lambda x: (d - x).days >= 0)
    is_before_end = pd.to_datetime(cholera['day_end']).apply(lambda x: (d - x).days <= 0)
    newdf[d] = is_after_start & is_before_end
```

```python
long = pd.melt(newdf, id_vars=['aid', 'funder_ref', 'daily_spend'])
long = long[long['value']==True].groupby(['funder_ref', 
                                          'variable'])['daily_spend'].sum().reset_index()
sns.lineplot(data=long, x='variable', y='daily_spend', hue='funder_ref')
plt.xticks(rotation=90)
plt.show()
```

```python
ddict, ddict_cholera, ddict_cholera_vax = [], [], []
for d in days_2017 + days_2018:
    is_after_start = pd.to_datetime(bysect['day_start']).apply(lambda x: (d - x).days >= 0)
    is_before_end = pd.to_datetime(bysect['day_end']).apply(lambda x: (d - x).days <= 0)
    ddict.append((d, 
                  np.sum(bysect.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.sum(bysect.loc[is_after_start & is_before_end,:]['daily_spend'].tolist()),
                  np.mean(bysect.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.mean(bysect.loc[is_after_start & is_before_end,:]['day_length'].tolist())))

    is_after_start = pd.to_datetime(cholera['day_start']).apply(lambda x: (d - x).days >= 0)
    is_before_end = pd.to_datetime(cholera['day_end']).apply(lambda x: (d - x).days <= 0)
    ddict_cholera.append((d, 
                  np.sum(cholera.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.sum(cholera.loc[is_after_start & is_before_end,:]['daily_spend'].tolist()),
                  np.mean(cholera.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.mean(cholera.loc[is_after_start & is_before_end,:]['day_length'].tolist())))
                  
    is_after_start = pd.to_datetime(cholera_vax['day_start']).apply(lambda x: (d - x).days >= 0)
    is_before_end = pd.to_datetime(cholera_vax['day_end']).apply(lambda x: (d - x).days <= 0)
    ddict_cholera_vax.append((d, 
                  np.sum(cholera_vax.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.sum(cholera_vax.loc[is_after_start & is_before_end,:]['daily_spend'].tolist()),
                  np.mean(cholera_vax.loc[is_after_start & is_before_end,:]['daily_budget'].tolist()),
                  np.mean(cholera_vax.loc[is_after_start & is_before_end,:]['day_length'].tolist())))
```

```python
db = pd.DataFrame(ddict, columns=['date', 
                                  'budget', 
                                  'spend', 
                                  'avg_daily_spend_per_activity', 
                                  'avg_duration_per_activity'])
db_cholera = pd.DataFrame(ddict_cholera, columns=['date', 
                                                  'budget', 
                                                  'spend', 
                                                  'avg_daily_spend_per_activity',
                                                  'avg_duration_per_activity'])
db_cholera_vax = pd.DataFrame(ddict_cholera_vax, columns=['date', 
                                                          'budget',
                                                          'spend',
                                                          'avg_daily_spend_per_activity',
                                                          'avg_duration_per_activity'])
```

```python
f, ax = plt.subplots(figsize=(10,9), nrows=3)
sns.lineplot(data=db, x='date', y='budget', label='budgeted', ax=ax[0])
sns.lineplot(data=db, x='date', y='spend', color='darkred', label='spent', ax=ax[0])
sns.lineplot(data=db_cholera, x='date', y='budget', label='budgeted', ax=ax[1])
sns.lineplot(data=db_cholera, x='date', y='spend', color='darkred', label='spent', ax=ax[1])
sns.lineplot(data=db_cholera_vax, x='date', y='budget', label='budgeted', ax=ax[-1])
sns.lineplot(data=db_cholera_vax, x='date', y='spend', color='darkred', label='spent', ax=ax[-1])
for a in ax:
    a.axvline(datetime(2018,1,1), linestyle='--', color='grey')
ax[0].set_title('Activities in Health, General, Basic Health, Water Supply & Sanitation')
for a in ax:
    a.legend(loc='lower right')
    a.ticklabel_format(style='plain', axis='y')
ax[1].set_title('Activities mentioning cholera')
for a in ax:
    a.set_ylabel('USD')
ax[2].set_title('Activities mentioning cholera + vaccinations')
plt.tight_layout()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10,9), nrows=3)
sns.lineplot(data=db, x='date', y='avg_duration_per_activity', ax=ax[0])
sns.lineplot(data=db_cholera, x='date', y='avg_duration_per_activity', ax=ax[1])
sns.lineplot(data=db_cholera_vax, x='date', y='avg_duration_per_activity', ax=ax[-1])
for a in ax:
    a.axvline(datetime(2018,1,1), linestyle='--', color='grey')
ax[0].set_title('Activities in Water Supply & Sanitation')
for a in ax:
    a.ticklabel_format(style='plain', axis='y')
ax[1].set_title('Activities mentioning cholera')
for a in ax:
    a.set_ylabel('days')
ax[2].set_title('Activities mentioning cholera + vaccinations')
plt.tight_layout()
plt.show()
```
